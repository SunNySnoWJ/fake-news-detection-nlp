```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
from google.colab import files

# Download NLTK data
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Load test data
df = pd.read_csv('AI_latest_news.csv')

# Preprocessing setup
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Clean the text and convert it into meaningful words
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words and len(w) > 2]
    return " ".join(tokens)

# Preprocess titles and content
def preprocess_test_data(df):
    df['clean_title'] = df['title'].apply(clean_text)

    if 'content' in df.columns:
        df['clean_text'] = df['content'].apply(clean_text)
        df['text_len'] = df['clean_text'].apply(lambda x: len(x.split()))
        df['text_sentiment'] = df['clean_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
    else:
        df['text_len'] = df['clean_title'].apply(lambda x: len(x.split()))
        df['text_sentiment'] = df['clean_title'].apply(lambda x: TextBlob(x).sentiment.polarity)

    # Optional date parsing
    if 'publishedAt' in df.columns:
        df['publish_date'] = pd.to_datetime(df['publishedAt'], errors='coerce')
        df['publish_year'] = df['publish_date'].dt.year
        df['publish_month'] = df['publish_date'].dt.month

    return df

# Apply preprocessing
df_test = preprocess_test_data(df)

# Save cleaned test data
df_test.to_csv('cleaned_AI_test_data.csv', index=False)
print("✅ Cleaned test data saved to: cleaned_AI_test_data.csv")
print(df_test[['clean_title', 'text_len', 'text_sentiment']].head())
files.download('cleaned_AI_test_data.csv')
```
